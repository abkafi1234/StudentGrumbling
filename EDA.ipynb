```python
import pandas as pd
from factor_analyzer import FactorAnalyzer
import matplotlib.pyplot as plt

# Load your data
# Variables
variables = [
    "Non-STEM courses significantly increase my academic workload. (Work Overload)",
    "Managing both STEM and non-STEM courses is challenging. (Work Overload)",
    "I am unsure how non-STEM courses contribute to my academic goals. (Role Ambiguity)",
    "I struggle to see the practical application of non-STEM subjects compared to STEM subjects. (Role Ambiguity)",
    "Balancing non-STEM and STEM coursework leaves me with less time for personal activities. (Work–Home Conflict)"
]

# Map Likert responses to numeric values
likert_map = {
    "Strongly disagree": 1,
    "Disagree": 2,
    "Neutral": 3,
    "Agree": 4,
    "Strongly agree": 5
}

numeric_data = data[variables].replace(likert_map)
# Initialize and fit FactorAnalyzer
fa = FactorAnalyzer(n_factors=2, rotation='varimax')
fa.fit(numeric_data)

# Get factor loadings
loadings = fa.loadings_
loadings_df = pd.DataFrame(loadings, index=variables, columns=['Factor 1', 'Factor 2'])
plt.figure(figsize=(8, 8))
plt.axhline(0, color='grey', linewidth=0.8)
plt.axvline(0, color='grey', linewidth=0.8)

for i, var in enumerate(variables):
    plt.scatter(loadings_df.iloc[i, 0], loadings_df.iloc[i, 1], color='blue')
    plt.text(loadings_df.iloc[i, 0]+0.02, loadings_df.iloc[i, 1]+0.02, var, fontsize=10)

plt.xlabel('Factor 1')
plt.ylabel('Factor 2')
plt.title('EFA Factor Loading Plot (Biplot)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
  # replace with your CSV path

# Survey questions
questions = [
    "I believe non-STEM courses are valuable despite their challenges. (Willingness)",
   "I am open to integrating lessons from non-STEM courses into my academic work.  (Willingness)",
    "I will likely apply the knowledge I gain from non-STEM courses in my personal or professional life. (Long-Term Adoption)"
]

# Short labels for figure
short_labels = ["Q1", "Q2", "Q3"]

# Likert scale order
likert_order = ["Strongly disagree", "Disagree", "Neutral", "Agree", "Strongly agree"]

# Colors for Likert levels
colors = {
    "Strongly disagree": "#d73027",
    "Disagree": "#fc8d59",
    "Neutral": "#ffffbf",
    "Agree": "#91bfdb",
    "Strongly agree": "#4575b4"
}

# Function to create diverging stacked bar chart
def plot_diverging_bar(question, ax):
    counts = data[question].value_counts().reindex(likert_order, fill_value=0)
    proportions = counts / counts.sum()
    
    # Negative, Neutral, Positive
    negative = proportions[["Strongly disagree", "Disagree"]].sum()
    positive = proportions[["Agree", "Strongly agree"]].sum()
    neutral = proportions["Neutral"]
    
    # Plot negative responses to the left
    ax.barh(0, -proportions["Disagree"], color=colors["Disagree"], edgecolor='white', label="Disagree")
    ax.barh(0, -proportions["Strongly disagree"], left=-proportions["Disagree"], color=colors["Strongly disagree"], edgecolor='white', label="Strongly disagree")
    
    # Plot neutral in center
    ax.barh(0, neutral, color=colors["Neutral"], edgecolor='white', label="Neutral")
    
    # Plot positive to the right
    ax.barh(0, proportions["Agree"], left=neutral, color=colors["Agree"], edgecolor='white', label="Agree")
    ax.barh(0, proportions["Strongly agree"], left=neutral+proportions["Agree"], color=colors["Strongly agree"], edgecolor='white', label="Strongly agree")
    
    # ax.set_xlim(-1, 1)  # Centered on 0
    ax.set_yticks([0])
    ax.set_yticklabels([""])
    ax.axvline(0, color='grey', linewidth=0.8)

# Create figure with 3 subplots
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 6), constrained_layout=True)

for i, question in enumerate(questions):
    plot_diverging_bar(question, axes[i])
    axes[i].set_ylabel(short_labels[i], rotation=0, labelpad=30, fontsize=12, va='center')

# Only show legend once
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right')

fig.suptitle("Figure 1: Distribution of Acceptance Responses", fontsize=16)
plt.show()
```

Justification for Sample Size in SEM (N≈200)
1. The Absolute Minimum Threshold Rule
Many foundational SEM sources suggest a minimum sample size of N=100 to N=200 for simple to moderately complex models. Your sample size of around 200 falls squarely within, or even exceeds, this common recommendation.

2. The Case-to-Parameter Ratio Rule (The Strongest Argument)
The most rigorous rule-of-thumb relates the sample size (N) to the number of free parameters (q) estimated in your model.

Your Model's Complexity: Your survey uses 14 Likert-scale items to measure 7 distinct latent constructs (Switching Cost, Benefit, Work Overload, Role Ambiguity, Grumbling, Willingness, and Adoption), plus several paths between them in a structural model. A typical SEM model built from this data would likely have between 20 and 35 free parameters (factor loadings, variances/covariances, and structural paths).

The Ratio:

The acceptable ratio is often cited as 10:1 (10 cases per parameter).

A more conservative ratio is 20:1 (20 cases per parameter).

10:1 (Acceptable)	10×30=300	211	Slightly Short/Borderline
5:1 (Bentler & Chou)	5×30=150	211	Adequate

Conclusion on Ratio: While you might be slightly below the conservative 10:1 ratio for a model with 30 parameters, you meet the widely-cited absolute minimum of N=200 and satisfy the 5:1 ratio. You can justify the sample size as adequate for detecting moderate to strong effects in a medium-complexity model.

3. Justification for Ordinal Data (The Crucial Caveat)
Since your data is based on 5-point Likert scales (ordinal data), the assumption of multivariate normality is violated. This makes the choice of the SEM estimator vital and directly impacts the required N.

Solution: You will use an estimator designed for ordinal data, such as Diagonally Weighted Least Squares (DWLS) or its variation, WLSMV (Weighted Least Squares Mean and Variance Adjusted), commonly used in Mplus.

The Justification: Using these robust estimators helps overcome the sample size limitations associated with non-normal data and makes the parameter estimates more stable. By explicitly stating that you used a robust estimator (e.g., WLSMV) appropriate for ordinal data, you effectively justify why N≈200 is sufficient, as these methods often perform well at this sample size for factor analysis and structural models.

In summary, you can confidently state that N≈200 is an acceptable sample size for your SEM analysis, particularly when employing an appropriate robust estimator for the ordinal nature of your Likert scale data.
